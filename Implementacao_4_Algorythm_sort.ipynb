#1) Insection Sort

```python
def insertion_sort(lista):

    num_testes = 0
    num_trocas = 0

    for i in range(1, len(lista)):
        chave = lista[i]
        j = i - 1

        while j >= 0:
            num_testes += 1

            if lista[j] > chave:
                lista[j + 1] = lista[j]
                num_trocas += 1
                j -= 1
            else:
                break

        lista[j + 1] = chave
        num_trocas += 1

    return lista, num_testes, num_trocas

if __name__ == "__main__":

    listas_teste = [
        [12, 42, 83, 25, 67, 71, 3, 4, 94, 53],
        [100, 48, 19, 61, 86, 33, 13, 43, 84, 28],
        [81, 60, 6, 49, 40, 41, 38, 64, 44, 36],
        [45, 27, 11, 89, 63, 39, 9, 58, 52, 17],
        [88, 77, 26, 62, 30, 96, 56, 65, 98, 99],
        [76, 73, 16, 95, 35, 87, 68, 69, 51, 92],
        [37, 75, 90, 82, 8, 18, 23, 93, 57, 10],
        [15, 97, 14, 29, 7, 24, 31, 59, 78, 85],
        [5, 70, 55, 91, 47, 72, 2, 20, 34, 74],
        [50, 66, 32, 22, 54, 79, 21, 1, 80, 46]
    ]

for i, lista_original in enumerate(listas_teste):

        lista_ordenada, testes, trocas = insertion_sort(lista_original.copy())

        print("\n" + "----------")
        print(f"Teste {i + 1}:")
        print(f"Lista Original: {lista_original}")
        print(f"Lista Ordenada: {lista_ordenada}")
        print(f"Contagem:")
        print(f"Quantidade de Testes : {testes}")
        print(f"Quantidade de Trocas : {trocas}")
```

#2) Selection Sort

```python
def selection_sort(lista):
    n = len(lista)
    num_testes = 0
    num_trocas = 0

    for i in range(n - 1):
        indice_minimo = i

        for j in range(i + 1, n):
            num_testes += 1
            if lista[j] < lista[indice_minimo]:
                indice_minimo = j

        if indice_minimo != i:
            lista[i], lista[indice_minimo] = lista[indice_minimo], lista[i]
            num_trocas += 1

    return lista, num_testes, num_trocas

if __name__ == "__main__":

  listas_teste = [
      [12, 42, 83, 25, 67, 71, 3, 4, 94, 53],
      [100, 48, 19, 61, 86, 33, 13, 43, 84, 28],
      [81, 60, 6, 49, 40, 41, 38, 64, 44, 36],
      [45, 27, 11, 89, 63, 39, 9, 58, 52, 17],
      [88, 77, 26, 62, 30, 96, 56, 65, 98, 99],
      [76, 73, 16, 95, 35, 87, 68, 69, 51, 92],
      [37, 75, 90, 82, 8, 18, 23, 93, 57, 10],
      [15, 97, 14, 29, 7, 24, 31, 59, 78, 85],
      [5, 70, 55, 91, 47, 72, 2, 20, 34, 74],
      [50, 66, 32, 22, 54, 79, 21, 1, 80, 46]
  ]

  for i, lista_original in enumerate(listas_teste):
          lista_ordenada, testes, trocas = selection_sort(lista_original.copy())

          print("\n" + "----------")
          print(f"Teste {i + 1}:")
          print(f"Lista Original: {lista_original}")
          print(f"Lista Ordenada: {lista_ordenada}")
          print(f"Contagem:")
          print(f"Quantidade de Testes : {testes}")
          print(f"Quantidade de Trocas : {trocas}")


```

#3) Bubblesort

```python
def bubble_sort(lista):

    n = len(lista)
    num_testes = 0
    num_trocas = 0

    for i in range(n - 1):
        troca_ocorreu = False

        for j in range(n - 1 - i):
            num_testes += 1

            if lista[j] > lista[j + 1]:
                lista[j], lista[j + 1] = lista[j + 1], lista[j]
                num_trocas += 1
                troca_ocorreu = True

        if not troca_ocorreu:
            break

    return lista, num_testes, num_trocas

if __name__ == "__main__":

    listas_teste = [
        [12, 42, 83, 25, 67, 71, 3, 4, 94, 53],
        [100, 48, 19, 61, 86, 33, 13, 43, 84, 28],
        [81, 60, 6, 49, 40, 41, 38, 64, 44, 36],
        [45, 27, 11, 89, 63, 39, 9, 58, 52, 17],
        [88, 77, 26, 62, 30, 96, 56, 65, 98, 99],
        [76, 73, 16, 95, 35, 87, 68, 69, 51, 92],
        [37, 75, 90, 82, 8, 18, 23, 93, 57, 10],
        [15, 97, 14, 29, 7, 24, 31, 59, 78, 85],
        [5, 70, 55, 91, 47, 72, 2, 20, 34, 74],
        [50, 66, 32, 22, 54, 79, 21, 1, 80, 46]
    ]

    for i, lista_original in enumerate(listas_teste):
        lista_ordenada, testes, trocas = bubble_sort(lista_original.copy())

        print("\n" + "----------")
        print(f"Teste {i + 1}:")
        print(f"Lista Original: {lista_original}")
        print(f"Lista Ordenada: {lista_ordenada}")
        print(f"Contagem:")
        print(f"Quantidade de Testes : {testes}")
        print(f"Quantidade de Trocas : {trocas}")
```

#4) Shell Sort

```python
def shell_sort(lista):

    n = len(lista)
    num_testes = 0
    num_trocas = 0

    gap = n // 2

    while gap > 0:
        for i in range(gap, n):
            chave = lista[i]
            j = i

            while j >= gap:
                num_testes += 1

                if lista[j - gap] > chave:
                    lista[j] = lista[j - gap]
                    num_trocas += 1
                    j -= gap
                else:

                    break

            lista[j] = chave
            num_trocas += 1


        gap //= 2

    return lista, num_testes, num_trocas

if __name__ == "__main__":

    listas_teste = [
        [12, 42, 83, 25, 67, 71, 3, 4, 94, 53],
        [100, 48, 19, 61, 86, 33, 13, 43, 84, 28],
        [81, 60, 6, 49, 40, 41, 38, 64, 44, 36],
        [45, 27, 11, 89, 63, 39, 9, 58, 52, 17],
        [88, 77, 26, 62, 30, 96, 56, 65, 98, 99],
        [76, 73, 16, 95, 35, 87, 68, 69, 51, 92],
        [37, 75, 90, 82, 8, 18, 23, 93, 57, 10],
        [15, 97, 14, 29, 7, 24, 31, 59, 78, 85],
        [5, 70, 55, 91, 47, 72, 2, 20, 34, 74],
        [50, 66, 32, 22, 54, 79, 21, 1, 80, 46]
    ]


    for i, lista_original in enumerate(listas_teste):
        lista_ordenada, testes, trocas = shell_sort(lista_original.copy())

        print("\n" + "----------")
        print(f"Teste {i + 1}:")
        print(f"Lista Original: {lista_original}")
        print(f"Lista Ordenada: {lista_ordenada}")
        print(f"Contagem:")
        print(f"Quantidade de Testes : {testes}")
        print(f"Quantidade de Trocas : {trocas}")
```

#5) Merge Sort

```python
def merge_sort(lista):

    contadores = {'testes': 0, 'trocas': 0}

    def merge(esquerda, direita):
        i = j = k = 0
        lista_final = [0] * (len(esquerda) + len(direita))

        while i < len(esquerda) and j < len(direita):
            contadores['testes'] += 1

            if esquerda[i] <= direita[j]:
                lista_final[k] = esquerda[i]
                i += 1
            else:
                lista_final[k] = direita[j]
                j += 1

            contadores['trocas'] += 1
            k += 1

        while i < len(esquerda):
            lista_final[k] = esquerda[i]
            contadores['trocas'] += 1
            i += 1
            k += 1

        while j < len(direita):
            lista_final[k] = direita[j]
            contadores['trocas'] += 1
            j += 1
            k += 1

        return lista_final

    def sort(lista):
        if len(lista) <= 1:
            return lista

        meio = len(lista) // 2

        esquerda = sort(lista[:meio])
        direita = sort(lista[meio:])

        return merge(esquerda, direita)

    lista_ordenada = sort(lista)

    return lista_ordenada, contadores['testes'], contadores['trocas']

if __name__ == "__main__":
    listas_teste = [
        [12, 42, 83, 25, 67, 71, 3, 4, 94, 53],
        [100, 48, 19, 61, 86, 33, 13, 43, 84, 28],
        [81, 60, 6, 49, 40, 41, 38, 64, 44, 36],
        [45, 27, 11, 89, 63, 39, 9, 58, 52, 17],
        [88, 77, 26, 62, 30, 96, 56, 65, 98, 99],
        [76, 73, 16, 95, 35, 87, 68, 69, 51, 92],
        [37, 75, 90, 82, 8, 18, 23, 93, 57, 10],
        [15, 97, 14, 29, 7, 24, 31, 59, 78, 85],
        [5, 70, 55, 91, 47, 72, 2, 20, 34, 74],
        [50, 66, 32, 22, 54, 79, 21, 1, 80, 46]
    ]

    for i, lista_original in enumerate(listas_teste):
        lista_ordenada, testes, trocas = merge_sort(lista_original.copy())

        print("\n" + "----------")
        print(f"Teste {i + 1}:")
        print(f"Lista Original: {lista_original}")
        print(f"Lista Ordenada: {lista_ordenada}")
        print(f"Contagem:")
        print(f"Quantidade de Testes : {testes}")
        print(f"Quantidade de Trocas : {trocas}")
```

#6) Quick Sort

```python
def quick_sort(lista):

    contadores = {'testes': 0, 'trocas': 0}

    def particiona(lista, inicio, fim):
        pivo = lista[fim]
        i = inicio - 1

        for j in range(inicio, fim):
            contadores['testes'] += 1
            if lista[j] <= pivo:
                i += 1
                lista[i], lista[j] = lista[j], lista[i]
                contadores['trocas'] += 1

        lista[i + 1], lista[fim] = lista[fim], lista[i + 1]
        contadores['trocas'] += 1

        return i + 1

    def quicksort_recursivo(lista, inicio, fim):
        if inicio < fim:
            indice_pivo = particiona(lista, inicio, fim)

            quicksort_recursivo(lista, inicio, indice_pivo - 1)
            quicksort_recursivo(lista, indice_pivo + 1, fim)

    quicksort_recursivo(lista, 0, len(lista) - 1)

    return lista, contadores['testes'], contadores['trocas']

if __name__ == "__main__":

    listas_teste = [
        [12, 42, 83, 25, 67, 71, 3, 4, 94, 53],
        [100, 48, 19, 61, 86, 33, 13, 43, 84, 28],
        [81, 60, 6, 49, 40, 41, 38, 64, 44, 36],
        [45, 27, 11, 89, 63, 39, 9, 58, 52, 17],
        [88, 77, 26, 62, 30, 96, 56, 65, 98, 99],
        [76, 73, 16, 95, 35, 87, 68, 69, 51, 92],
        [37, 75, 90, 82, 8, 18, 23, 93, 57, 10],
        [15, 97, 14, 29, 7, 24, 31, 59, 78, 85],
        [5, 70, 55, 91, 47, 72, 2, 20, 34, 74],
        [50, 66, 32, 22, 54, 79, 21, 1, 80, 46]
    ]

    for i, lista_original in enumerate(listas_teste):

        lista_ordenada, testes, trocas = quick_sort(lista_original.copy())

        print("\n" + "----------")
        print(f"Teste {i + 1}:")
        print(f"Lista Original: {lista_original}")
        print(f"Lista Ordenada: {lista_ordenada}")
        print(f"Contagem:")
        print(f"Quantidade de Testes : {testes}")
        print(f"Quantidade de Trocas : {trocas}")
```

#7) Heap Sort

```python
def heap_sort(lista):

    n = len(lista)
    contadores = {'testes': 0, 'trocas': 0}

    def heapify(lista, n, i):
        maior = i
        esquerda = 2 * i + 1
        direita = 2 * i + 2

        if esquerda < n:
            contadores['testes'] += 1
            if lista[esquerda] > lista[maior]:
                maior = esquerda

        if direita < n:
            contadores['testes'] += 1
            if lista[direita] > lista[maior]:
                maior = direita

        if maior != i:
            lista[i], lista[maior] = lista[maior], lista[i]
            contadores['trocas'] += 1

            heapify(lista, n, maior)

    for i in range(n // 2 - 1, -1, -1):
        heapify(lista, n, i)

    for i in range(n - 1, 0, -1):
        lista[i], lista[0] = lista[0], lista[i]
        contadores['trocas'] += 1

        heapify(lista, i, 0)

    return lista, contadores['testes'], contadores['trocas']

if __name__ == "__main__":

    listas_teste = [
        [12, 42, 83, 25, 67, 71, 3, 4, 94, 53],
        [100, 48, 19, 61, 86, 33, 13, 43, 84, 28],
        [81, 60, 6, 49, 40, 41, 38, 64, 44, 36],
        [45, 27, 11, 89, 63, 39, 9, 58, 52, 17],
        [88, 77, 26, 62, 30, 96, 56, 65, 98, 99],
        [76, 73, 16, 95, 35, 87, 68, 69, 51, 92],
        [37, 75, 90, 82, 8, 18, 23, 93, 57, 10],
        [15, 97, 14, 29, 7, 24, 31, 59, 78, 85],
        [5, 70, 55, 91, 47, 72, 2, 20, 34, 74],
        [50, 66, 32, 22, 54, 79, 21, 1, 80, 46]
    ]


    for i, lista_original in enumerate(listas_teste):
        lista_ordenada, testes, trocas = heap_sort(lista_original.copy())

        print("\n" + "----------")
        print(f"Teste {i + 1}:")
        print(f"Lista Original: {lista_original}")
        print(f"Lista Ordenada: {lista_ordenada}")
        print(f"Contagem:")
        print(f"Quantidade de Testes: {testes}")
        print(f"Quantidade de Trocas: {trocas}")
```

#8) Counting Sort

```python
def counting_sort(lista):

    n = len(lista)
    contadores = {'testes': 0, 'trocas': 0}

    valor_maximo = 100
    vetor_contagem = [0] * (valor_maximo + 1)
    vetor_saida = [0] * n

    for num in lista:
        vetor_contagem[num] += 1
        contadores['testes'] += 1

    for i in range(1, valor_maximo + 1):
        vetor_contagem[i] += vetor_contagem[i - 1]
        contadores['testes'] += 1

    for i in range(n - 1, -1, -1):
        num = lista[i]

        posicao = vetor_contagem[num] - 1
        vetor_saida[posicao] = num
        contadores['trocas'] += 1

        vetor_contagem[num] -= 1
        contadores['testes'] += 1

    return vetor_saida, contadores['testes'], contadores['trocas']

if __name__ == "__main__":

    listas_teste = [
        [12, 42, 83, 25, 67, 71, 3, 4, 94, 53],
        [100, 48, 19, 61, 86, 33, 13, 43, 84, 28],
        [81, 60, 6, 49, 40, 41, 38, 64, 44, 36],
        [45, 27, 11, 89, 63, 39, 9, 58, 52, 17],
        [88, 77, 26, 62, 30, 96, 56, 65, 98, 99],
        [76, 73, 16, 95, 35, 87, 68, 69, 51, 92],
        [37, 75, 90, 82, 8, 18, 23, 93, 57, 10],
        [15, 97, 14, 29, 7, 24, 31, 59, 78, 85],
        [5, 70, 55, 91, 47, 72, 2, 20, 34, 74],
        [50, 66, 32, 22, 54, 79, 21, 1, 80, 46]
    ]

    for i, lista_original in enumerate(listas_teste):
        lista_ordenada, testes, trocas = counting_sort(lista_original.copy())

        print("\n" + "----------")
        print(f"Teste {i + 1}:")
        print(f"Lista Original: {lista_original}")
        print(f"Lista Ordenada: {lista_ordenada}")
        print(f"Contagem:")
        print(f"Quantidade de Testes : {testes}")
        print(f"Quantidade de Trocas : {trocas}")
```
